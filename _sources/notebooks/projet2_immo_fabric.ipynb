  <!-- TEXTE -->
  <p>
    Ce projet s’appuie sur les données DVF publiques (plus de 20 millions de lignes) afin d’analyser
    le marché immobilier français. Les données sont traitées à travers une architecture en médaillon 
    (bronze → silver → gold), nettoyées et modélisées sous forme de schéma en étoile 
    combinant faits et dimensions.  
    L’ensemble est exploité dans Microsoft Fabric pour produire un rapport Power BI 
    mettant en avant les tendances temporelles et géographiques du marché.
  </p>
