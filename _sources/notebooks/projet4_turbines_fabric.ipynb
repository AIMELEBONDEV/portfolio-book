  <!-- TEXTE -->
  <p style="text-align: justify;">
    Ce projet vise à traiter, analyser et monitorer les données d’un parc de trois éoliennes.  
    Les données sont collectées toutes les dix minutes depuis des capteurs, et nous disposons également 
    d’historiques au format CSV. L’ensemble des données est intégré dans Microsoft Fabric via plusieurs 
    Lakehouses selon l’architecture en médaillon : bronze (données brutes), silver (données nettoyées et enrichies) 
    et gold (données prêtes à l’analyse).  
  </p>

  <p style="text-align: justify;">
    Les données sont traitées grâce à deux méthodes : des Dataflows pour une approche sans code, et des Notebooks 
    (SQL/Python) pour les transformations plus avancées. Une fois les données consolidées et structurées 
    sous forme de schéma en étoile, elles sont utilisées pour créer un modèle sémantique et générer des rapports 
    Power BI. Ces rapports permettent de suivre la production d’électricité de chaque turbine et d’alerter en cas 
    d’anomalies, par exemple via Teams ou Outlook.  
  </p>

  <p style="text-align: justify;">
    L’architecture est entièrement automatisée à l’aide de pipelines, orchestrant la récupération quotidienne 
    des données, leur traitement et la mise à jour des rapports. Ce projet offre une vision complète des étapes 
    indispensables pour maîtriser Microsoft Fabric, de l’ingestion de données jusqu’au monitoring automatisé.
  </p>
