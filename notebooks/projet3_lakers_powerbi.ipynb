  <p>
    Passionné par le basketball et supporter des <strong>Los Angeles Lakers</strong>, 
    j’ai conçu ce projet pour suivre et analyser les performances de la franchise.  
    Les données sont extraites directement depuis l’<em>API officielle de la NBA</em>, 
    couvrant les équipes, les joueurs et les matchs sur 5 saisons (2020 à 2025).  
    La base de données est conçue pour s’actualiser automatiquement au fil du temps 
    (par exemple en 2026 et au-delà).
  </p>

  <p>
    L’objectif principal est de comprendre ce qui contribue aux victoires des Lakers, 
    à travers l’étude des <em>statistiques collectives de l’équipe</em> (points, passes décisives, 
    rebonds, interceptions, contres, turnovers) et des <em>performances individuelles des joueurs</em>.  
    Le projet combine une architecture de collecte et de stockage robuste avec une couche d’analyse 
    interactive (rapports et visualisations).
  </p>
