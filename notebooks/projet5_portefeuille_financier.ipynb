  <p style="text-align: justify;">
    Ce projet consiste à analyser les performances d’un portefeuille financier en utilisant les données récupérées 
    via l’API de <em>Yahoo Finance</em>. Les données concernent les indices de marché, comme le S&P 500, ainsi que 
    les données journalières des actions individuelles.  
  </p>

  <p style="text-align: justify;">
    À partir de ces données, plusieurs indicateurs financiers et techniques sont calculés afin de mesurer la 
    performance du portefeuille, identifier les actifs les plus performants et ceux nécessitant un ajustement, 
    et guider la prise de décision. Parmi les indicateurs calculés, on retrouve notamment le retour journalier, 
    le retour moyen, ainsi que les moyennes mobiles pour l’analyse technique.  
  </p>

  <p style="text-align: justify;">
    L’ensemble du projet a été réalisé en <strong>Python</strong>, permettant à la fois l’extraction des données 
    de l’API et le traitement analytique pour générer des insights sur le portefeuille. 
    Les résultats ont également été exploités dans <strong>Power BI</strong> pour créer des rapports interactifs, 
    visualiser les performances et suivre l’évolution du portefeuille dans le temps.  
    Ce projet fournit ainsi un outil complet pour évaluer, analyser et optimiser un portefeuille financier.
  </p>
